```python
def clean_gender(gender):
    gender = gender.lower().strip()

    malekeywords = ['m','male','male-ish','maile','something kinda male?','cis male','mal','male (cis)','make','guy (-ish) ^_^','male leaning androgynous','man','msle','mail','malr','cis man','ostensibly male, unsure what that really means']
    femalekeywords = ['female','cis female','f','woman','queer/she/they','femake','cis-female/femme','female (cis)','femail']
    transkeywords = ['trans-female','trans woman','female (trans)','all']
    nonbinarykeywords = ['non-binary','nah','enby','fluid','genderqueer','androgyne','agender','neuter','queer','p','a little about']

    for word in malekeywords:
        if word in gender:
            return "Male"
    
    for word in femalekeywords:
        if word in gender:
            return "Female"
        
    for word in transkeywords:
        if word in gender:
            return 'Transgender'
        
    for word in nonbinarykeywords:
        if word in gender:
            return 'Non Binary'
    
```

```python
df['Gender'] = df['Gender'].apply(clean_gender)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['treatment'],axis = 1)
y = df['treatment']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
